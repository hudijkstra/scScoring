<h3> Notebook contains code for preprocessing eQTL data for use in PascalX as mapperfile and BED file for open chromatin intersection.

```python
import pandas as pd
from pathlib import Path
```

```python
DATA_DIR = Path("../Data/PascalX/resources/")
eqtl_file = DATA_DIR / "sceqtlgen_all_sig_eqtls.tsv.gz"
ref_file = DATA_DIR / "1000G-30x.ref.gz"
gene_ref = DATA_DIR / "limix_annotation_strand.tsv"
```

```python
# Load eQTL data
eqtl_df = pd.read_csv(eqtl_file, sep="\t", compression="gzip")
eqtl_df
```

```python
# Load reference panel
ref_df = pd.read_csv(ref_file, compression="gzip", delim_whitespace=True)
ref_df
```

```python
merged_df = eqtl_df.merge(
    ref_df,
    left_on=['snp_chromosome', 'snp_position'],
    right_on=['CHR', 'bp'],
    how='inner'
)
merged_df
```

```python
# Filter HLA region
merged_df = merged_df[~(
    (merged_df["feature_chromosome"] == 6) &
    (merged_df["feature_start"] >= 25_000_000) &
    (merged_df["feature_end"] <= 35_000_000)
)]
merged_df
```

```python
mapper_df = merged_df[["ID", "ENSG", "feature_id", "empirical_feature_p_value"]].rename(columns={
    "ENSG": "ENSG",
    "ID": "rsID",
    "feature_id": "gene_symbol",
    "empirical_feature_p_value" : "pval" 
})

mapper_df
```

```python
gr_df = pd.read_csv(gene_ref, sep="\t")
print(gr_df)
# Find the overlap between ENSG values in both DataFrames
overlap = mapper_df["ENSG"].isin(gr_df["ENSG"])

# Count how many overlap
num_overlap = overlap.sum()

print(f"Number of overlapping ENSG entries: {num_overlap}")
```

```python
mapper_file = DATA_DIR / "gwas_eqtl_intersect.tsv"
mapper_df.to_csv(mapper_file, sep="\t", index=False)
```

<h2> Create BED file for intersecting with open chromatin</h2>

```python
bed_dir = "../Data/ATAC/unmapped_snps"
bed_file = bed_dir + "/eqtls.bed"

merged_df["snp_position+1"] = (merged_df["snp_position"] + 1).copy()
bed_df = merged_df[[ "snp_chromosome", "snp_position", "snp_position+1", "ID",  "ENSG", "feature_id"]]

bed_df["snp_chromosome"] = bed_df["snp_chromosome"].astype(int)
bed_df["snp_position"] = bed_df["snp_position"].astype(int)
bed_df["snp_position+1"] = bed_df["snp_position+1"].astype(int)

bed_df.to_csv(bed_file, sep="\t", index=False, header=False)
```
